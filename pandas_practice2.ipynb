```python
import pandas as pd
```

```python
data = { "name" : ['A','B','C','D','E','F','G'],
         "horse_power": [130,250,190,300,210,220,170],
          "weight" : [1.9,2.6,2.2,2.9,2.4,2.3,2.2],
          "efficiency" : [16.3,10.2,11.1,7.1,12.1,13.2,14.2]}
          
```

```python
df = pd.DataFrame(data)
```

```python
df = df.set_index('name',drop=True)
```

```python
df
```

```python
max_result = '마력과 연비를 곱한 값중 최대값을 가진 차종 : ' + df.index[df['horse_power']*df['efficiency'] == (df['horse_power']*df['efficiency']).max()]
```
